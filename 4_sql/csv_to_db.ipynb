```python
import pandas as pd
from sqlalchemy import create_engine
from google.colab import files
```

```python
df = pd.read_csv("/content/dataset_final.csv")
```

```python
#cria o banco
engine = create_engine('sqlite:///projeto_final.db')

#salva todas as tabelas nesse banco que foi criado
df.to_sql('netflix_titles',con=engine, if_exists='replace', index=False)

print("Banco de dados criado com sucesso!")
```
